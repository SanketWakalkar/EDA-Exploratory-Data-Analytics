```python
# import your code here
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
bmi_scaled = SS.fit_transform(data[['bmi']])
print(np.round(bmi_scaled.var(),2), np.round(bmi_scaled.mean(),2))
print(bmi_scaled.min(), bmi_scaled.max())
```
